```python
import cv2
import os
import matplotlib.pyplot as plt
from IPython.display import display, Image
import numpy as np
os.environ["TF_CPP_MIN_LOG_LEVEL"] = '2'
file_name = "1003-receipt.jpg"
file_full_path = os.path.join(os.getcwd(),"large-datasets",file_name)
# Read the image
image = cv2.imread(file_full_path)

# Convert the image to grayscale
grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
plt.imshow(grayscale_image)
```

```python
noise_reduced_image = cv2.GaussianBlur(grayscale_image, (5, 5), 0)
plt.imshow(noise_reduced_image)
```

```python
clahe = cv2.createCLAHE(clipLimit=10.0, tileGridSize=(8, 8))
contrast_enhanced_image = clahe.apply(grayscale_image)
plt.imshow(contrast_enhanced_image)
```

```python
_, binarized_image = cv2.threshold(grayscale_image, 127, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Display the binarized image on the notebook
plt.imshow(binarized_image, cmap='Greys_r')
plt.show()
```

```python
kernel = np.ones((4, 4), np.uint8)  # You can adjust the kernel size

# Erosion followed by dilation is a common morphological operation (opening)
opened_image = cv2.morphologyEx(binarized_image, cv2.MORPH_OPEN, kernel)

# Display the result after noise removal
plt.imshow(opened_image, cmap='Greys_r')
plt.show()
```

```python
edges = cv2.Canny(opened_image, 50, 150)

# Display the edge-detected image on the notebook
plt.imshow(edges, cmap='Greys_r')
plt.show()
```

```python
lines = cv2.HoughLines(edges, 1, np.pi/180, threshold=100)

# Find the angle of the most dominant line
angle = lines[0][0][1] if lines is not None and len(lines) > 0 else 0
angle_degrees = np.degrees(angle)

# Rotate the image to straighten it
rows, cols = opened_image.shape
rotation_matrix = cv2.getRotationMatrix2D((cols/2, rows/2), angle_degrees, 1)
straightened_image = cv2.warpAffine(opened_image, rotation_matrix, (cols, rows))

# Display the straightened image
plt.imshow(straightened_image, cmap='Greys_r')
plt.show()
```
